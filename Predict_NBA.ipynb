```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/data/nba_games.csv",index_col=0)
```

```python
df
```

```python
df=df.sort_values("date")
```

```python
df=df.reset_index(drop=True)
```

```python
df
```

```python
df.columns
```

```python
# deleting duplicated columns
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
# on the right side we can see a column win , we want to now add a column called "target" so that we can set it as our "y" later on, this is what our model is going to predict
def add_target(team):
  team["target"] = team["won"].shift(-1)
  return team

df = df.groupby("team", group_keys= False).apply(add_target)
```

```python
# we can see that a column "target" has been added and its nothing but the result of the game after the current one
df[df["team"]=="CLE"]
```

```python
df["target"][pd.isnull(df["target"])] = 2
# reason we have null values is because there were no games played after that
```

```python
# converting bool to int
df["target"] = df["target"].astype(int,errors="ignore")
```

```python
df
```

```python
# dropping null columns
df = df.dropna(axis=1,how="all")
```

```python
df.head()
```

```python
from sklearn.model_selection import TimeSeriesSplit # during cross validation we want to verify with date suceeding the training set datemeaning
# meaning we want to predict for new date from old date instead of predicting games of 2015 using 2020 data
from sklearn.feature_selection import SequentialFeatureSelector # Trains the model using different sets of features (columns) because 142 columns is a lot and might cause overfitting
from sklearn.linear_model import RidgeClassifier # its a classifier model

cv= TimeSeriesSplit(n_splits=3) # 3 indicates the data will train on 2 and cross validate on 1
model=RidgeClassifier(alpha=1) # alpha is similar to learning rate, increasing is what makes it different from other classifiers
sfs=SequentialFeatureSelector(model,n_features_to_select=30,cv=cv,direction="forward")
```

```python
# we are gonna scale (normalize) the data
from sklearn.preprocessing import MinMaxScaler
# I dont want to impute mean values on NaN in target so I will not be using SimpleImputer
removed_columns=["season","date","won","target","team","team_opp"]
# apple minmaxscaler on numerical columns
scaler=MinMaxScaler()
selected_columns=df.columns[~df.columns.isin(removed_columns)]
df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
df
```

```python
X=df[selected_columns]
y= df["target"]
```

```python
# Applying SequentialFeatureSelector to find the best features for making the actual model
sfs.fit(X,y)
```

```python
predictors = list(selected_columns[sfs.get_support()])
```

```python
predictors
```

```python
list(df["season"].unique())
# we are gonna be using minimum two years of historic data to predict new data
# eg: 2016,2017 will be used to predict 2018 season
# 2016,2017,2018 will be used to predict 2019 season
# etc
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = []

    seasons = sorted(data["season"].unique())

    for i in range(start, len(seasons), step):
        season = seasons[i]
        train = data[data["season"] < season]
        test = data[data["season"] == season]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(df,model,predictors)
```

```python
sorted(df["season"].unique())
```

```python
predictions
```

```python
# checking accuracy of model
from sklearn.metrics import accuracy_score
accuracy_score(predictions["actual"],predictions["prediction"])
# 54% - model is basically guessing (realistically this is pretty good for a model that predicts a game)
# predicting a game is different from predicting a spam email, the latter is 100% bound to be spam whereas anything can happen in a match
```

```python
# In most sports many matches may just have been played differently one after the other, therefore we will take average of 10 games for better fitting
```

```python
df_rolling = df[list(selected_columns) + ["won","season","team"]]
```

```python
# creating rows that find average of the past 10 rows
def find_team_average(team):
  rolling = team.rolling(10).mean()
  return rolling

df_rolling=df_rolling.groupby(["team","season"],group_keys=False).apply(find_team_average)
```

```python
# renaming columns
rolling_cols = [f"{cols}_10" for cols in df_rolling.columns]

df_rolling.columns = rolling_cols
```

```python
# giving data we know before hand about the game
# in the real world we know wheter the team is playing at home and we know against whom
# giving the model a little bit of information for better judgement
```

```python
def shift_col(team,col_name):
  new_col = team[col_name].shift(-1)
  return new_col

def add_col(df, col_name):
  return df.groupby("team",group_keys=False).apply(lambda x : shift_col(x , col_name))
```
